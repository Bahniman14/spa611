## Q1. Verify that the mean of power is centered around the variance of the voltage

```python
import numpy as np

# Parameters
N = 1000000
sigma = 2.0   # RMS

# Generate Gaussian voltage
V = np.random.normal(0, sigma, N)

# Compute statistics
variance_voltage = np.var(V)
mean_power = np.mean(V**2)

print("RMS (sigma):", sigma)
print("Theoretical Variance (sigma^2):", sigma**2)
print("Computed Variance:", variance_voltage)
print("Mean Power:", mean_power)
print("Difference:", abs(mean_power - variance_voltage))
```

#### Explaination : The diffrence between Mean Power and Volate Varinece is very small. So we can say that the power is centered around the varience of volatage.

## Q2. Consider cases of different RMS and plot their histograms. [5 marks]

```python
import numpy as np
import matplotlib.pyplot as plt

N = 500000
sigmas = [0.5, 1.0, 2.0, 2.5, 3.0, 3.5]

plt.figure(figsize=(10,6))

for sigma in sigmas:
    V = np.random.normal(0, sigma, N)
    plt.hist(V, bins=200, density=True, alpha=0.5, label=f"RMS = {sigma}")

plt.title("Voltage Histogram for Different RMS")
plt.xlabel("Voltage")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()
plt.show()
```

#### Explaination: As we increase the RMS the Histogram widens, Variance increases but the Mean remains zero

## Q3. How will the histogram change if I add a slowly varying sine wave in it

```python
import numpy as np
import matplotlib.pyplot as plt

N = 500000
sigma = 1.0
A = 2.0
t = np.linspace(0, 100*np.pi, N)

# Gaussian noise
noise = np.random.normal(0, sigma, N)

plt.figure(figsize=(10,6))
plt.hist(noise, bins=200, density=True, alpha=0.7)
plt.title("Histogram: Noise")
plt.xlabel("Voltage")
plt.ylabel("Probability Density")
plt.grid()
plt.show()

# Slowly varying sine wave
signal1 = A * np.sin(0.001 * t)

# Combined signal
V = noise + signal1

plt.figure(figsize=(10,6))
plt.hist(V, bins=200, density=True, alpha=0.7)
plt.title("Histogram: Noise + Slowly Varying Sine [Amp: 0.01]")
plt.xlabel("Voltage")
plt.ylabel("Probability Density")
plt.grid()
plt.show()

print("Mean Voltage:", np.mean(V))
print("Variance:", np.var(V))

signal2 = A * np.sin(0.01 * t)

# Combined signal
V = noise + signal2

plt.figure(figsize=(10,6))
plt.hist(V, bins=200, density=True, alpha=0.7)
plt.title("Histogram: Noise + Slowly Varying Sine [Amp : 0.1]")
plt.xlabel("Voltage")
plt.ylabel("Probability Density")
plt.grid()
plt.show()

print("Mean Voltage:", np.mean(V))
print("Variance:", np.var(V))

```

#### Explaination: We get a Gaussian whose mean shifts slowly with time

## Q4. How are the Variances of the ensembles distributed ? [5 marks]

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000          # Samples per ensemble
M = 10000         # Number of ensembles
sigma = 1.5

variance_estimates = []

for _ in range(M):
    V = np.random.normal(0, sigma, N)
    variance_estimates.append(np.var(V))

variance_estimates = np.array(variance_estimates)

plt.figure(figsize=(10,6))
plt.hist(variance_estimates, bins=100, density=True)
plt.title("Distribution of Variance Estimates Across Ensembles")
plt.xlabel("Estimated Variance")
plt.ylabel("Probability Density")
plt.grid()
plt.show()

print("Theoretical Variance:", sigma**2)
print("Mean of Estimated Variances:", np.mean(variance_estimates))
print("Standard Deviation of Estimates:", np.std(variance_estimates))
```

#### Explaination: The variances of the ensembles follow a scaled chi-square distribution centered around the true variance, which becomes approximately Gaussian for large sample sizes.
